```python
#%% import required libraries
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig
%matplotlib inline
```

Create a random signal X with length x_n

```python
x_n = 15
np.random.seed(0)
X = np.random.randn(x_n)
plt.plot(X)
#print x
```

Stuff X at n_t random locations in a zero array H with length H_n

```python
n_t = 5 
H_n = 500
taps = np.random.randint(low = 0 ,high = H_n-x_n,size=(n_t))
H = np.zeros((H_n))
for t in taps:
    H[t:(t+x_n)] = X
plt.plot(H)
```

replace zeros with random numbers

```python
#Y = sig.convolve(X,H,mode = 'full')
np.random.seed(1)
Y = H + np.random.randn(len(H))*(H == 0)
plt.plot(Y)
```

H*(x) =  X(-x)

```python
#H_star = map(lambda x : 1.0/x,np.flipud(X))
H_star = np.flipud(X)
y2 = sig.convolve(Y,H_star)
plt.plot(y2.flatten())
```

All the patterns are detected using matched filter kernel
